### Background
**Kinematics** is not dynamics, it is just a *description* of how 2D vector fields vary in 2D. 

* Assume LINEAR dependence of u,v on x,y 
    * This involves 4 coefficients $c_1, c_2, c_3, c_4$
* Also, add a constant (not variying with x,y) vector field U,V 

$$ u = U + c_1 x  +c_2 y$$
$$ v = V + c_3 x  +c_4 y$$

## Exercise 1 Answer:

Considering that
* $ u = U + c_1 x  +c_2 y$
* $ v = V + c_3 x  +c_4 y$
* $\frac{\partial u}{\partial x}=c_1$
* $\frac{\partial u}{\partial y}=c_2$
* $\frac{\partial v}{\partial x}=c_3$
* $\frac{\partial v}{\partial y}=c_4$

From table 7.1 (p.272): 
* $ div  = \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y}$
* $ \zeta_{rel}  = \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y}$
* $ def1 = \frac{\partial u}{\partial x} - \frac{\partial v}{\partial y}$
* $ def2 = \frac{\partial v}{\partial x} + \frac{\partial u}{\partial y}$

Therefore:
* $ div = \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y}= \frac{\partial (U + c_1 x  +c_2 y)}{\partial x} + \frac{\partial (V + c_3 x  +c_4 y)}{\partial y} = c_1+c_4$
* $ \zeta_{rel} = \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y} = \frac{\partial (V + c_3 x  +c_4 y)}{\partial x} - \frac{\partial (U + c_1 x  +c_2 y)}{\partial y} = c_3-c_2$
* $ def1 = \frac{\partial u}{\partial x} - \frac{\partial v}{\partial y} = \frac{\partial (U + c_1 x  +c_2 y)}{\partial x} - \frac{\partial (V + c_3 x  +c_4 y)}{\partial y} = c_1-c_4$
* $ def2 = \frac{\partial v}{\partial x} + \frac{\partial u}{\partial y} = \frac{\partial (V + c_3 x  +c_4 y)}{\partial x} + \frac{\partial (U + c_1 x  +c_2 y)}{\partial y} = c_3+c_2$

## Exercise 2: Create and display the basic fields
* First, execute a code cell setting up a coordinate system (x,y) from -10 to 10
* Next, execute the cell defining the basic "pure" div, vort, def1, def2, and U,V fields 
* Finally, display them. The first one has been done for you. You complete the work. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up a coordinate grid centered on 0,0
n = 10
x,y = np.mgrid[-n:n+1, -n:n+1]
```

```python
# Pure divergence of unit magnitude
udiv,vdiv = 1*x, 1*y

# Pure vorticity of unit magnitude
uvor,vvor = -1*y, 1*x

# Pure def1 of unit magnitude
udef1,vdef1 = 1*x, -1*y

# Pure def2 of unit magnitude
udef2,vdef2 = 1*y, 1*x

# Pure U and V
uU, vU = 1, 0
uV, vV = 0, 1
```

```python
# Define the display field udisp,vdisp and give it a label. 
# Edit-> Copy and paste this cell, then adjust the top 2 lines for the other cases

#### PURE DIVERGENCE ####
udisp,vdisp = udiv, vdiv
label = 'Pure unit Divergence'

# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()

#### PURE VORTICITY ####
udisp,vdisp = uvor, vvor
label = 'Pure unit Vorticity'

# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()

#### PURE DEF1 ####
udisp,vdisp = udef1, vdef1
label = 'Pure unit Deformation 1'

# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()

#### PURE DEF2 ####
udisp,vdisp = udef2, vdef2
label = 'Pure unit Deformation 2'

# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()
```

## Answer to Exercise 3 (vor +def1): 

Vorticity plus deformation1 (two curved flow patterns) makes a flow we could describe as NE/SW orientated flow. It creates a pattern with 2 opposite fluxes (one above another). A line of no movement is located along y=x (NE/SW) line. Vectors on the right (left) of the no-motion line have northeastern (southwestern) flow. As farther from the no-motion line, greater the intensity of the flow.
